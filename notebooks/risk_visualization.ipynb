```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the processed dataset
df = pd.read_csv("C:\\Users\\HP\\week 6\\Credit-Scoring-Model\\data\\processed_data1.csv")

# Count plot for risk categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Risk', palette='viridis')
plt.title('Distribution of Risk Categories')
plt.xlabel('Risk Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Pie chart for risk categories
risk_counts = df['Risk'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(risk_counts, labels=risk_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', n_colors=len(risk_counts)))
plt.title('Risk Categories Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
